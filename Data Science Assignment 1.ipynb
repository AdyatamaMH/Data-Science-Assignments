```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

# Downloading CNBCI data and CNNI

headers = {'Accept-Language': 'en-US,en;q=0.8'}
url = 'https://www.cnbcindonesia.com/news'
url2 = 'https://en.antaranews.com/'
response = requests.get(url,headers=headers)
response2 = requests.get(url2,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
soup2 = BeautifulSoup(response2.text, "html.parser")


newstitle = soup.select('div a h2')
print(newstitle[0])
print(" ")
print("how many titles are available?",len(soup.select('div a h2')))

newstitle2 = soup2.select('header h3')
print(newstitle2[0])
print("")
print("how many titles are available?",len(soup2.select('header h3')))

text = soup.select('li article a')
links=[]
for a in soup.select('li article a'):
  links.append(a.attrs.get('href'))

text2 = soup2.select('header h3 a')
links2= []
for b in soup2.select('header h3 a'):
    links2.append(b.attrs.get('href'))

print(links[0])
print(links2[0])
```

```python
import requests
from bs4 import BeautifulSoup

article_title = []
article_links = []
author = []
summary = []
date = []

article_title2 = []
article_links2 = []
author2 = []
summary2 = []
quote2 = []


for x in links:
    url_n1 = x
    response_n1 = requests.get(url_n1, headers=headers)
    soup_n1 = BeautifulSoup(response_n1.text, "html.parser")

    author_tag = soup_n1.find('div', class_='author')
    summary_tag = soup_n1.find('div', class_='media_artikel')
    date_tag = soup_n1.find('div', class_='date')

    if author_tag:
        author.append(author_tag.get_text(strip=True))
    else:
        author.append(None)

    if summary_tag:
        summary.append(summary_tag.get_text(strip=True))
    else:
        summary.append(None)

    if date_tag:
        date.append(date_tag.get_text(strip=True))
    else:
        date.append(None)
        
for z in links2:
    url2_n2 = z
    response2_n2 = requests.get(url2_n2, headers=headers)
    soup2_n2 = BeautifulSoup(response2_n2.text, "html.parser")
    
    author2_tag2 = soup2_n2.find('p', class_='text-muted small mt10') 
    summary2_tag2 = soup2_n2.find('p', class_='wp-caption-text')
    quote2_tag2 = soup2_n2.find('div', class_='quote_old')
    
    if author2_tag2:
        author2.append(author2_tag2.get_text(strip=True))
    else:
        author2.append(None)

    if summary2_tag2:
        summary2.append(summary2_tag2.get_text(strip=True))
    else:
        summary2.append(None)

    if quote2_tag2:
        quote2.append(quote2_tag2.get_text(strip=True))
    else:
        quote2.append(None)


        

for t in soup.select('div a h2'):
    article_title.append(t.get_text())

for link in soup.select('li article a'):
    article_links.append(link.attrs.get('href'))
    
for c in soup2.select('header h3'):
    article_title2.append(c.get_text())

for link in soup2.select('header h3 a'):
    article_links2.append(link.attrs.get('href'))

print(article_title)
print(article_links)
print(author)
print(summary)
print(date)

print(article_title2)
print(article_links2)
print(author2)
print(summary2)
print(quote2)
```

```python
df = pd.DataFrame(
        {'article title': article_title, 'links': article_links, 'author': author, 'Picture': summary, 'date' : date
}
    )

df2 = pd.DataFrame(
    {'article_title2': article_title2, 'article_links2': article_links2, 'author2': author2, 'Summary': summary2, 'quote2' : quote2}
)


print (df.head())
print (df2.head())


df.to_csv('CNBCI.csv', index=False)
df2.to_csv('ANTARNEWS.csv', index=False)
```
